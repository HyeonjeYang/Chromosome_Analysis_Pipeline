extract coordinates from .h5 file to .csv file\
you can use this file to render at blender.\
I used blender 5.0

```python
import numpy as np
import h5py
import os
```

```python
h5_path = "trajectory_20260222_185839/blocks_0-99.h5" #path example, you can change the path according to your file name
```

```python
f = h5py.File(h5_path, "r")
print(list(f.keys()))
```

You can see that each frame has its own unique coordinates

```python
print(list(f["0"].keys()))
```

```python
save_folder = "frames_npy"
os.makedirs(save_folder, exist_ok=True)
with h5py.File(h5_path, "r") as f:
    

    frame_keys = sorted(f.keys(), key=lambda x: int(x))
    
    for i, key in enumerate(frame_keys):
        coords = f[key]["pos"][:] 
        np.save(f"{save_folder}/frame_{i:04d}.npy", coords) #you can also save as .csv

print("Saved all frames.")
```
